```python
# Import Libraries in the Notebook
import pandas as pd
import numpy as np
import geopandas as gpd
import folium
from folium.plugins import HeatMap
from shapely.geometry import Point, Polygon
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

# Load the citibike data
citibike_data = pd.read_csv('data\citibike_clean.csv')
```

```python
# Convert start and end coordinates into geospatial points
citibike_data['start_geometry'] = citibike_data.apply(lambda row: Point(row['start_station_longitude'], row['start_station_latitude']), axis=1)
citibike_data['end_geometry'] = citibike_data.apply(lambda row: Point(row['end_station_longitude'], row['end_station_latitude']), axis=1)

# Convert the pandas dataframe into a geospatial dataframe
citibike_data_start = gpd.GeoDataFrame(citibike_data, geometry='start_geometry')
citibike_data_end = gpd.GeoDataFrame(citibike_data, geometry='end_geometry')

# Display the first few rows of the geospatial dataframe
citibike_data_start.head()
```
